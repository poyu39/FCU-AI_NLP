```python
import pandas as pd

df = pd.read_csv('./dataset/online_shopping_10_cats.csv')
print(df.head())
```

```python
# 丟掉 cat 欄位 (只需要訓練分類label)

df = df.drop(columns=['cat'])

print(df.head())
```

```python
# jieba 斷詞
import jieba

def cut_review(review):
    review = str(review)
    return ' '.join(jieba.cut(review))

# df = pd.DataFrame(df['label'], df['review'].astype(str))
df['review'] = df['review'].apply(cut_review)

print(df.head())
```

```python
# 清除標點符號
import re

def remove_punctuation(text):
    return re.sub(r'[^\w\s]', '', text)

df['review'] = df['review'].apply(remove_punctuation)

# 把多個空白變成一個空白
def replace_multi_space(text):
    return re.sub(r'\s+', ' ', text)

df['review'] = df['review'].apply(replace_multi_space)

print(df.head())
```

```python
# 清除停用詞
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')

stop_words = stopwords.words('chinese')

print(stop_words[:10])

def remove_stop_words(text):
    return ' '.join([word for word in text.split() if word not in stop_words])

df['review'] = df['review'].apply(remove_stop_words)

print(df.head())
```

```python
# TFIDF 向量化

from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=2500, min_df=7, max_df=0.8)

tfidfconverter = vectorizer.fit_transform(df['review']).toarray()
```

```python
# 切割訓練集與測試集
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(tfidfconverter, df['label'], test_size=0.2, random_state=0)

print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Decision Tree 訓練模型

from sklearn.tree import DecisionTreeClassifier

DT_text_classifier = DecisionTreeClassifier()
DT_text_classifier.fit(X_train, y_train)
```

```python
# 預測
DT_predictions = DT_text_classifier.predict(X_test)
```

```python
# 分數
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(y_test, DT_predictions))
print(accuracy_score(y_test, DT_predictions))
print(classification_report(y_test, DT_predictions))
```

```python
# DL
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=X_train.shape[1]))
model.add(Dense(units=6, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
model.summary()
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, batch_size=64, epochs=10, validation_data=(X_test, y_test))
```

```python
# score
score = model.evaluate(X_test, y_test)
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# 混淆矩陣
import numpy as np
y_pred = model.predict(X_test)
y_pred = np.round(y_pred)
print(confusion_matrix(y_test, y_pred))
```
